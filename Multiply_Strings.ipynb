```python

    

def simple_mult(str1, str2): 
    multiplications_dict = {}

    for i in range(0, 10): 
        multiplications_dict[str(i)]={}
        for j in range(0, 10): 
            multiplications_dict[str(i)][str(j)] = str(i*j)
    
    answer = multiplications_dict[str1][str2]
    
    carry = '0' 
    
    if len(answer)==1:
        carry='0'
    
    if len(answer)==2: 
        carry=answer[0]
        
    
    return (carry, answer[-1])



def dict_sum(a,b): 
    sum_dict={}

    for i in range(0, 10):
        sum_dict[str(i)]={}
        for j in range(0, 10):
            sum_dict[str(i)][str(j)]=str(i+j)
    
            
    return sum_dict[a][b]
    



def mult_row_and_char(a, b): 
    
    if len(a)==1 and len(b)==1: 
        carry = str(simple_mult(a,b)[0])
        result = str(simple_mult(a,b)[1])
        
        result_list= [result]
        if carry=='0': 
            return result_list
        else: 
            result_list.insert(0, carry)
            return result_list
        
    
    a_reverse = a[::-1]
    # print("a_reverse:{}".format(a_reverse))
    
    # print(b[-1])
    current_char=b[-1]
    
    carry = '0'
    
    row = []
    
    i = 0 
    while i < len(a_reverse): 
        result = dict_sum(simple_mult(current_char, a_reverse[i])[1], carry)
        carry = str(simple_mult(current_char, a_reverse[i])[0])
        
        row.append(result)
        # print("{}) {}x{}={}, carry={}".format(i, current_char, a_reverse[i], result, carry))
        i=i+1
        
        
    if carry!='0':
        row.insert(0, str(carry))
        
    
    row.reverse()

    return row
    
          

            
            
            
    
          



        

# print(dict_sum('6','2'))


# print(mult_row_and_char('123', '6'))

# print(mult_row_and_char('123', '5'))



a = '123'
b = '456' 

def get_rows(a,b): 
    b_reverse = b[::-1]

    rows = []

    for i in range(0, len(b)): 
        result_row =mult_row_and_char(a, b_reverse[i])
        if i>0: 
            zeros_to_append = list(i*'0')
            result_row = result_row+zeros_to_append

        rows.append(result_row)
        # print(mult_row_and_char(a, b_reverse[i]))
        
    max_digits = 0

    for i in range(0, len(rows)): 
        if len(rows[i])>max_digits: 
            max_digits= len(rows[i])

    # print(max_digits)

    for i in range(0, len(rows)): 
        num_zeros = max_digits - len(rows[i])
        rows[i] = list(num_zeros*'0')+rows[i]
        
    return rows 

rows= get_rows('123','456')

# print(rows)

def column_sum(rows): 
    temp_sums = []
    temp_sums_items = []
    for i in range(len(rows[0])-1,-1, -1): 
        temp_sum = '0'
        temp_sums_items.append(temp_sum)
        for j in range(0, len(rows)): 
            a= temp_sum
            b = rows[j][i] 
            temp_sum = dict_sum(a,b)
        temp_sums.append(temp_sum)
        temp_sums_items.append(temp_sum)
    #print("temp_sums_items:{}".format(temp_sums_items))
    return temp_sums
    

# print(column_sum(rows))

column_sums = column_sum(rows)

def get_final_output(a,b): 
    
    rows= get_rows(a,b)
    column_sums = column_sum(rows)


    final_output = []
    carry = '0'

    for i in range(0, len(column_sums)):
        if len(column_sums[i]) == 1:
            # print("len is 1, i:{}, column_sums[i] ={}, carry:{}".format(i, column_sums[i], carry ))
            final_output.append(dict_sum(carry,column_sums[i][-1]))
        else: 
            final_output.append(dict_sum(carry,column_sums[i][-1]))
            carry = column_sums[i][0]
            # print("len is > 1, i:{}, column_sums[i] ={}, carry:{}".format(i, column_sums[i], carry ))

    final_output.reverse()
    
    final_output = ''.join(map(str, final_output))
    
    return final_output
        
        
        
print(get_final_output('1365', '17')) 
    
    

    
                
                
                
                


        
```
