```python
class MyCircularQueue:

    def __init__(self, k: int):
        self.queue = ['']*k 
        

    def enQueue(self, value: int) -> bool:
        i = self.first_empty_index()
        
        if i ==-1: 
            return False 
        try: 
            i = self.first_empty_index()

            self.queue[i] = value 
            return True 
        except: 
            return False 
        

    def deQueue(self) -> bool:
        i = self.first_non_empty_index()
            
        if i ==-1: 
            return False 
            
        try: 
            i = self.first_non_empty_index()
            print("first_non_empty_index:{}".format(i))
            self.queue[i]=''
            print("first_non_empty_index:{}".format(i))

            
            a = self.first_non_empty_index()
            b = self.last_non_empty_index()
            
            non_empty_list= self.queue[a:b+1]
            print("non_empty_list:{}".format(non_empty_list))
            
            remaining_spaces= ['']*(len(self.queue) -len(non_empty_list))
            
            self.queue = non_empty_list + remaining_spaces
                                    
            
            return True 
        except: 
            return False
            


    def Front(self) -> int:
        for i in range(0, len(self.queue)):
            if self.queue[i]!='': 
                return self.queue[i]
        else:
            return -1
        

    def Rear(self) -> int:
        for x in self.queue[::-1]: 
            if x!='': 
                return x 
        else: 
            return -1 

        

    def isEmpty(self) -> bool:
        count = 0 
        
        for i in range(0, len(self.queue)): 
            if self.queue[i] =='':
                count+=1 
        
        if count == len(self.queue): 
            return True 
        else: 
            return False 
        

    def isFull(self) -> bool:
        count = 0 
        
        for i in range(0, len(self.queue)): 
            if self.queue[i] !='':
                count+=1 
        
        if count == len(self.queue): 
            return True 
        else: 
            return False 
    
    def first_empty_index(self) -> int: 
        empty_index = -1 
        for i in range(0, len(self.queue)): 
            if self.queue[i]=='': 
                return i 
            
        return empty_index 
    
    def first_non_empty_index(self) -> int: 
        non_empty_index = -1 
        for i in range(0, len(self.queue)): 
            if self.queue[i]!='': 
                return i 
            
        return non_empty_index 
    
    def last_non_empty_index(self) -> int: 
        non_empty_index = -1 
        
        for i in range(0, len(self.queue)): 
            if self.queue[::-1][i]!='': 
                return (len(self.queue)-i-1) 

        return non_empty_index 
    
    
```
