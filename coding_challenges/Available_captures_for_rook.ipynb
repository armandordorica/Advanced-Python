```python
board = [[".",".",".",".",".",".",".","."],[".",".",".","p",".",".",".","."],[".",".",".","R",".",".",".","p"],[".",".",".",".",".",".",".","."],[".",".",".",".",".",".",".","."],[".",".",".","p",".",".",".","."],[".",".",".",".",".",".",".","."],[".",".",".",".",".",".",".","."]]


# board = [[".",".",".",".",".",".",".","."],[".","p","p","p","p","p",".","."],[".","p","p","B","p","p",".","."],[".","p","B","R","B","p",".","."],[".","p","p","B","p","p",".","."],[".","p","p","p","p","p",".","."],[".",".",".",".",".",".",".","."],[".",".",".",".",".",".",".","."]]


board = [[".",".",".",".",".",".",".","."],[".",".",".","p",".",".",".","."],[".",".",".","p",".",".",".","."],["p","p",".","R",".","p","B","."],[".",".",".",".",".",".",".","."],[".",".",".","B",".",".",".","."],[".",".",".","p",".",".",".","."],[".",".",".",".",".",".",".","."]]


board = [[".",".",".",".",".",".",".","."],[".",".",".","p",".",".",".","."],[".",".",".","R",".",".",".","p"],[".",".",".",".",".",".",".","."],[".",".",".",".",".",".",".","."],[".",".",".","p",".",".",".","."],[".",".",".",".",".",".",".","."],[".",".",".",".",".",".",".","."]]


board = [[".",".",".",".",".",".",".","."],[".","p","p","p","p","p",".","."],[".","p","p","B","p","p",".","."],[".","p","B","R","B","p",".","."],[".","p","p","B","p","p",".","."],[".","p","p","p","p","p",".","."],[".",".",".",".",".",".",".","."],[".",".",".",".",".",".",".","."]]


def visualize_board(board): 
    for i in range(0, len(board)): 
        print(board[i])
        
# visualize_board(board)

def find_rook(board): 
    for i in range(0, len(board)): 
        for j in range(0, len(board[i])): 
            if board[i][j]=='R': 
                return (i,j)
            
def find_pawns(board): 
    pawns_locations = []
    for i in range(0, len(board)): 
        for j in range(0, len(board[i])): 
            if board[i][j]=='p': 
                pawns_locations.append((i,j))
    
    return pawns_locations


def find_north_pawns(board): 
    northern_pawns = 0
    rook_location= find_rook(board)
    rook_i = rook_location[0]
    rook_j = rook_location[1]
    
    # print(rook_i, rook_j)
    
    for i in range(rook_i, -1, -1): 
        # print(i,rook_j)
        # print(board[i][rook_j])
        if board[i][rook_j] == 'p':
            northern_pawns +=1
            return northern_pawns 
        if board[i][rook_j] == 'B':
            return 0 
    return 0 
        
        
def find_south_pawns(board): 
    southern_pawns = 0
    rook_location= find_rook(board)
    rook_i = rook_location[0]
    rook_j = rook_location[1]
    
    # print(rook_i, rook_j)
    
    for i in range(rook_i+1, len(board)): 
        # print(i,rook_j)
        # print(board[i][rook_j])
        if board[i][rook_j] == 'p':
            southern_pawns +=1
            return southern_pawns 
        if board[i][rook_j] == 'B':
            return 0 
    return 0


def find_east_pawns(board): 
    eastern_pawns = 0
    rook_location= find_rook(board)
    rook_i = rook_location[0]
    rook_j = rook_location[1]
    
    # print(rook_i, rook_j)
    
    for j in range(rook_j+1, len(board)): 
        # print(rook_i,j)
        # print(board[i][rook_j])
        if board[rook_i][j] == 'p':
            eastern_pawns +=1
            return eastern_pawns 
        if board[rook_i][j] == 'B':
            return 0 
    return 0 
    

def find_west_pawns(board): 
    western_pawns = 0
    rook_location= find_rook(board)
    rook_i = rook_location[0]
    rook_j = rook_location[1]
    
    # print(rook_i, rook_j)
    
    for j in range(rook_j-1,-1, -1): 
        # print(rook_i,j)
        # print(board[rook_i][j])
        if board[rook_i][j] == 'p':
            western_pawns +=1
            return western_pawns 
        if board[rook_i][j] == 'B':
            return 0 
    return 0 


            
                

visualize_board(board)
# print(find_rook(board))
# print(find_pawns(board))
# print(find_north_pawns(board))
# print(find_south_pawns(board))

# print(find_east_pawns(board))

print(find_west_pawns(board))

def get_num_avail_captures(board): 
    return find_north_pawns(board)+find_south_pawns(board)+find_east_pawns(board)+find_west_pawns(board)


print(get_num_avail_captures(board))


```
