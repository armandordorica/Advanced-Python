```python


units_dict = {'0':'Zero',
             '1':'One', '2': 'Two', '3':'Three', 
            '4':'Four', '5':'Five', '6':'Six', 
            '7':'Seven', '8':'Eight', '9':'Nine', 
             '10':'Ten', '11':'Eleven', '12':'Twelve', 
             '13':'Thirteen', '14':'Fourteen', '15':'Fifteen', 
             '16':'Sixteen', '17':'Seventeen', '18':'Eighteen',
             '19':'Nineteen'}

tens_dict = {'20':'Twenty', '30':'Thirty', '40':'Forty', 
            '50':'Fifty', '60':'Sixty', '70':'Seventy', 
            '80':'Eighty', '90':'Ninety'}

length_to_units = { '3':'Hundred', '4':'Thousand', '7':'Million', '10':'Billion'}



def integer_to_words(x): 
    

    if x<=19: 
        output_string=units_dict[str(x)]
        return output_string 

    if len(str(x))==2: #20-99 
        x_str = str(x)
        tens = str(int(x_str[0])*10)
        first_word = tens_dict[tens]

        second_word = units_dict[x_str[1]]
        output_string = first_word + ' ' + second_word 
        
        return output_string 
    
    if len(str(x)) >=3 and len(str(x))<4:
        base = 100
        first_word = integer_to_words(x//base) + " " + length_to_units['3']
        
        remainder = x - x//base *base
        second_word = integer_to_words(remainder) 
        
        if second_word == 'Zero':
            second_word = ''
        
        output_string = first_word + ' ' + second_word 
        
        return output_string 
    
    if len(str(x)) >=4 and len(str(x))<7:
        base = 1000
        first_word = integer_to_words(x//base) + " " + length_to_units['4']
        
        remainder = x - x//base *base
        second_word = integer_to_words(remainder) 
        
        if second_word == 'Zero':
            second_word = ''
        
        output_string = first_word + ' ' + second_word 
        
        return output_string 
    
    if len(str(x)) >=7 and len(str(x))<10:
        base = 1000000
        first_word = integer_to_words(x//base) + " " + length_to_units['7']
        
        remainder = x - x//base *base
        second_word = integer_to_words(remainder) 
        
        if second_word == 'Zero':
            second_word = ''
        
        output_string = first_word + ' ' + second_word 
        
        return output_string 
    
    if len(str(x)) >=10 and len(str(x))<12:
        base = 1000000000
        first_word = integer_to_words(x//base) + " " + length_to_units['10']
        
        remainder = x - x//base *base
        second_word = integer_to_words(remainder) 
        
        if second_word == 'Zero':
            second_word = ''
        
        output_string = first_word + ' ' + second_word 
        
        return output_string 
        
        
        
    
#     if len(str(x)) >=4 and len(str(x))<7: 
#         base = 1000
#         first_word = 
        #Thousands 
        
#     if len(str(x)) >=7 and len(str(x))<10: 
#         #Millions 
        
#     if len(str(x)) >=10 and len(str(x))<12: 
#         #Billions
        
#     if len(str(x)) >=12 and len(str(x))<15: 
#         #Trillions 
  
input = 1234567891
print(str(input) +": " + integer_to_words(input))
        

    
# for i in range(0, len(first_100_nums)): 
#     print(first_100_nums[i], integer_to_words(first_100_nums[i]))
                                                            
    
    
    
    
    




    
```
